```python
# import data from results
import pickle 
import numpy as np
with open('results/pareto_success_noisy.pkl', 'rb') as f:
    success_noisy = pickle.load(f)
with open('results/pareto_success_bayes.pkl', 'rb') as f:
    success_bayes = pickle.load(f)
eps_list = np.linspace(0.1, 1, 5)
alpha = 0.05
```

```python
success_noisy
```

```python
success_bayes
```

```python
import matplotlib.pyplot as plt
# make color list
plt.errorbar(eps_list, success_noisy[:, 0], yerr=[success_noisy[:, 1], success_noisy[:, 2]], fmt='--', label=f"Noisy Binary Search")
plt.errorbar(eps_list, success_bayes[:, 0], yerr=[success_bayes[:, 1], success_bayes[:, 2]], fmt='-', label=f"Bayesian Screening Search")
    
plt.xlabel('privacy budget', fontsize=14)
plt.xticks(fontsize=12)
plt.ylabel('Success Rate', fontsize=14)
plt.yticks(fontsize=12)
# plot legend outside on the right
plt.legend()
plt.title(f'Success Rate vs. privacy budget (alpha 0.05), coins 10K, users 5k, num_exp 200', fontsize=10)
plt.savefig('results/pareto_success.png', bbox_inches='tight')
plt.show()
```
